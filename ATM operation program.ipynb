```python
class Bank:
    
    name="Manoj Narware"
    Acc_no=68001947535
    def __init__(self):
        self.bal=0
        self.pin=1234

    def deposit(self,dep):
        if dep<0:
            return "Invalid Amount !!!"
        else:
            self.bal=self.bal+dep
            return "{} Rs. has been deposited successfully!!!".format(dep)

    def withdraw(self,withd):
        if withd>self.bal:
            return "Insufficient Balance !!!"
        elif withd<0:
            return "Invalid Amount!!!"
        else:
            self.bal=self.bal-withd
            return"{} Rs. has been debited!!!".format(withd)
    def balinfo(self):
        return("Your current bal is :{} Rs.".format(self.bal))
    def setPin(self,pin):
        if len(str(pin))==4:
            if self.pin==pin:
                return "Please do not set previous PIN,set different!!!"
            else:
                self.pin=pin
                return "Your PIN has been successfully changed !!!"
        else:
            return"PIN digit should be four!!!"
            
            
```
